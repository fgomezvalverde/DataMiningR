## 1.1 Contexto inicial

Las estadísticas de accidentes automovilísticas en el transcurso del años son alarmantes, estos han enlutado a muchos hogares costarricenses, y son una de las principales causas de muerte de este país; son lamentables las estadísticas reportadas del promedio de muertes al día. 

La problemática relacionada con los accidentes de tránsito en Costa Rica alcanza niveles dramáticos, no solo para las familias que pierden un ser querido, sino para aquellos niños adolescentes y adultos, que sobreviven a un accidente de tránsito, pero quedan con secuelas graves y permanentes, que les impiden disfrutar de una vida plena en el futuro.

Los accidentes en su mayoría se pueden prevenir, ya que los factores condicionantes se asocian a ingesta de alcohol, imprudencia de los peatones, adelantamiento indebido, imprudencia del conductor, imprudencia del ciclista, falla mecánica, poca visibilidad en las carreteras, exceso de velocidad y a las malas condiciones de la carreteras, en las que intervienen la mala iluminación, el ancho de los carriles, la mala señalización, y las condiciones climáticas de la zona. 

Muchos accidentes de tránsito se pueden evitar o mitigar sus consecuencias, mediante la aplicación de técnicas de la ingeniería vial. En Costa Rica se ha identificado por ejemplo la incorrecta colocación de barandas, muy poca señalización, elemento que si se corrige, es una medida muy barata que puede prevenir muchos accidentes. Tal es el caso de rótulos que indiquen: ir más despacio, peligro, curva adelante, altos, etc. También se ha probado que las señales de tránsito existentes, han perdido la calidad de reflexión que pueden tener en las noches. 

El estado en su afán por disminuir la inseguridad en las vías, está constantemente buscando hacer un frente común para actuar contra un problema de salud pública que figura como la primera causa de muerte prematura de la población costarricense. La situación, afecta a varias instituciones como: La Caja de Seguro Social, Mopt, Cosevi, INS y servicios de Emergencias y Urgencia, para estas instituciones es importante realizar intervenciones que tengan como propósito reducir las profundas secuelas que dejan estos percances en la sociedad costarricense.

Durante el 2019, en una conferencia el doctor Román Macaya Hayes, presidente ejecutivo de la Caja Costarricense de Seguro Social (CCSS), explicó que para la institución, los accidentes de tránsito constituyen uno de los problemas más serios que se atienden en los servicios de Emergencias y Urgencias y que desplaza a otros pacientes que requieren una consulta, un procedimiento, una cirugía, o una atención en los servicios de Emergencias.

Para este proyecto se analizará el problema con el fin de descubrir patrones sobre la problemática de los accidentes de tránsito en Costa Rica, para discutir sobre nuevas acciones que se podrían emprender para ponerle freno a la violencia que se vive en carreteras.

## 4.3 Selección de las técnicas a utilizar

### 4.3.1 Construcción del modelo - Random Forest

Los siguientes son los pasos básicos involucrados en la realización del algoritmo de bosque aleatorio:

* Eligir N estimadores aleatorios del conjunto de datos.
* Construir N cantidad de arboles , según la variable **n_estimators**.
* Determinar la capacidad máxima de profundidad de cada árbol, según la variable **max_depth**.
* Para este problema de clasificación, cada árbol en el bosque predice la categoría a la que pertenece el nuevo registro. Finalmente, el nuevo registro se asigna a la categoría que gana el voto mayoritario.

Random Forest es un método de aprendizaje conjunto que ajusta a múltiples árboles de decisión en subconjuntos de datos y promedia los resultados.

#### 4.3.1.1 Selección de los parámetros

Para este algoritmo los parametros seleccionados son:

*   **n_estimators:** para el caso de este parámetro vamos hacer un análisis para encontrar el mejor.

*   **max_depth:** le vamos asignar el valor **None**, lo que significa que cada árbol se expandirá hasta que cada hoja sea pura.

#### 4.3.1.2 Ejecución

```
vDecisionTreeResultEstimators = pd.DataFrame(columns = ['n_estimators' , 'Score',]) 
for estimator in range(1,100,10):
    vModelDecisionTreeEstimators =  RandomForestClassifier(n_estimators=estimator,max_depth=None,random_state = vRandomSeed)
    vModelDecisionTreeEstimators.fit(X_Train, Y_Train)
    vModelDecisionTreeEstimators_prediction = vModelDecisionTreeEstimators.predict(X_Test)
    vDecisionTreeResultEstimators = vDecisionTreeResultEstimators.append({'n_estimators' : estimator , 'Score' : accuracy_score(vModelDecisionTreeEstimators_prediction, Y_Test)} , ignore_index=True)
```

```
vDecisionTreeResultEstimators.plot(title = 'Evaluacion de Precisión según n_estimators', x ='n_estimators',y='Score',figsize=(vSeabornWidth ,vSeabornHeight))
```

Con esto definimos que la mejor configuración para el **Random Forest** es un **n_estimators** de **51**.

```
vModelDecisionTree = RandomForestClassifier(n_estimators=51,max_depth=None,random_state = vRandomSeed)
vModelDecisionTree.fit(X_Train, Y_Train)
vScore = vModelDecisionTree.score(X_Test, Y_Test)
print("La precisión alcanzada por Random Forest es de {0}%".format(vScore))
```

#### 4.3.1.3 Descripción de los modelos obtenidos

```
fig = plt.gcf()
fig.set_size_inches(  vSeabornBigWidth, vSeabornHeight)
feature_imp = pd.Series(vModelDecisionTree.feature_importances_,index=X_Train.columns).sort_values(ascending=False)
sns.barplot(x=feature_imp, y=feature_imp.index)
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.legend()
plt.show()
```

### 4.3.2 Construcción del modelo - Logistic Regression

La regresión logística es un algoritmo de clasificación utilizado para asignar observaciones a un conjunto discreto de clases. La regresión logística es un algoritmo de análisis predictivo que se basa en el concepto de probabilidad.

![texto alternativo](https://miro.medium.com/max/1400/1*PQ8tdohapfm-YHlrRIRuOA.gif)

#### 4.3.2.1 Selección de los parámetros

Para este algoritmo los parametros seleccionados son:
*   **solver:** Se seleccionó ’lbfgs’ que es el default y se utiliza para que el algoritmo optimice el problema y este maneja la pérdida multinomial.


*   **multi_class:** Se seleccionó ‘auto’ que es el valor default, pero que a la vez se adapta de acuerdo al tipo de problema.

Es decir, en lugar de elegir las siguientes opciones:
* "ovr", se ajusta un problema binario para cada etiqueta. 
* "liblinear" admite la regularización de L1 y L2. 
* "multinomial", la pérdida minimizada es el ajuste de pérdida multinomial en toda la distribución de probabilidad, incluso cuando los datos son binarios. 

Seleccionamos "Auto", ya que este selecciona "ovr" o "liblinear" si los datos son binarios, y de lo contrario selecciona "multinomial".

#### 4.3.2.2 Ejecución

```
vModeLogisticRegression= LogisticRegression(random_state=vRandomSeed, solver='lbfgs', multi_class='auto').fit(X_Train, Y_Train)
vModeLogisticRegression.predict(X_Test)
```

#### 4.3.2.3 Descripción de los modelos obtenidos

```
print("La precisión alcanzada por Logistic Regression es de {0}%".format(vModeLogisticRegression.score(X_Test,Y_Test)))
```

### 4.3.3 Construcción del modelo - K Neareast Neighbors Classifier

KNN es un algoritmo de aprendizaje no paramétrico y vago. No paramétrico significa que no existe un supuesto para la distribución de datos subyacente. 

En KNN, K es el número de vecinos más cercanos. El número de vecinos es el factor decisivo central. K es generalmente un número impar si el número de clases es 2. Cuando K = 1, entonces el algoritmo se conoce como el algoritmo vecino más cercano. Este es el caso más simple. Supongamos que P1 es el punto, para el cual la etiqueta necesita predecir. Primero, encuentra el punto más cercano a P1 y luego la etiqueta del punto más cercano asignado a P1.

![texto alternativo](http://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1531424125/Knn_k1_z96jba.png)

Supongamos que P1 es el punto, para el cual la etiqueta necesita predecir. Primero, encuentra el k punto más cercano a P1 y luego clasifica los puntos por mayoría de votos de sus k vecinos. Cada objeto vota por su clase y la clase con más votos se toma como la predicción.

#### 4.3.3.1 Selección de los parámetros

Para este algoritmo no se va a elegir solo un k, si no que vamos a realizar iteraciones para luego seleccionar el k que no brinde el mejor resultado.

En este caso, la iteración que vamos a hacer es de k = 1 hasta k = 10

#### 4.3.3.2 Ejecución

```
vRMSE_Values_KNeighbors = []
for k in range(1,10):
    vKNeighborsModel =  KNeighborsClassifier(n_neighbors=k)
    vKNeighborsModel.fit(X_Train, Y_Train)
    vKNeighborsModel_prediction = vKNeighborsModel.predict(X_Test)


    vPrediction = accuracy_score(vKNeighborsModel_prediction, Y_Test)
    vRMSE_Values_KNeighbors.append(vPrediction)
```

#### 4.3.3.3 Descripción de los modelos obtenidos

```
vCurve_KNeighbors = pd.DataFrame(vRMSE_Values_KNeighbors)
vCurve_KNeighbors.plot(title = 'Evaluacion de Precisión K Nearest Neighbors según K',figsize=(vSeabornWidth ,vSeabornHeight))
```
